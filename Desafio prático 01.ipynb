Atividade 03. Crie uma fun√ß√£o area(r, pi), que calcule a √°rea de um c√≠rculo.
A √°rea de um c√≠rculo √© dada por ùëé = ùúãùëü
2
, onde:
‚Ä¢ ùúã √© o n√∫mero pi, cujo valor √© ~3,14159265359
‚Ä¢ ùëü √© o raio do c√≠rculo
Esta fun√ß√£o deve receber dois argumentos, r e pi, e retornar a √°rea de um c√≠rculo
com raio r. Entretanto, voc√™ deve fornecer duas op√ß√µes de utiliza√ß√£o da fun√ß√£o:
a) A utiliza√ß√£o de um valor padr√£o para pi (ùúã = 3,14), sem a necessidade de
declarar explicitamente o seu valor;
b) A utiliza√ß√£o com a declara√ß√£o explicita do valor exato de pi desejado.
Por exemplo, para um raio ùëü = 8, o usu√°rio poder√° utilizar a sua fun√ß√£o das
seguintes maneiras:

Utiliza√ß√£o da fun√ß√£o sem declarar pi
Neste caso o valor de pi dever√° ser 3.14 e o resultado: 3.14 * (8 ** 2)
area(8)
Utiliza√ß√£o da fun√ß√£o declarando explicitamente o valor de pi
Neste caso o valor de pi dever√° ser 3,141592 e o resultado: 3,141592 * (8 ** 2)
area(8, 3,141592)
DICA: Para resolver esse problema, utilize o conceito de argumentos padr√µes (default).



```python
def calcular_area(r,pi=3.141592):
    return pi*(r*2)
```

```python
print(calcular_area(8))
print(calcular_area(8,3.14))
```

Atividade 05. Crie um m√≥dulo em Python, chamado processalista (arquivo processalista.py),
com as seguintes fun√ß√µes, que dever√£o receber como argumento uma lista (tipo
list) de n√∫meros inteiros

```python
# Encontra e retorna o maior n√∫mero impar presente na lista
def maior_impar(lista):
    maiores_impares = [valor for valor in lista if valor % 2 != 0]
    if maiores_impares:
        return max(maiores_impares)
    else:
        return None
            
# Encontra e retorna o menor n√∫mero impar presente na lista
def menor_impar(lista):
    menores_impares = [valor for valor in lista if valor % 2 != 0]
    if menores_impares:
        return min(menores_impares)
    else:
        return None
# Encontra e retorna o maior e o menor n√∫mero √≠mpar presentes na lista
def menor_maior_impar(lista):
    maior_impar(lista)
    menor_impar(lista)


```
